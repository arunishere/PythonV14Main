```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

## Line Chart

```python
df = pd.DataFrame()

df['Date'] = pd.date_range(start='2024-01-01', end='2024-12-30', freq='D')
df['Price'] = np.random.randint(320, 350, size=(365))
df['Price_2'] =  [2*x for x in range(0,365)]
df['Price_3'] =  [1.5*x for x in range(0,365)]
df['Price_4'] =  [1.75*x for x in range(0,365)]

df
```

```python
#Plotting a line.

plt.figure(figsize=(15,4))
plt.plot(df['Date'], df['Price_2'])
plt.plot(df['Date'], df['Price_3'])
plt.plot(df['Date'], df['Price_4'], linestyle='--')
plt.title('Sample Line Plotting')
plt.xlabel('Date')
plt.ylabel('Price($)')
plt.grid(True)
plt.legend(['2x', '1.5x', '1.75x'])
plt.show()

```

```python
plt.figure(figsize=(15,4))
# plt.plot(df['Date'], df['Price_2'])
# plt.plot(df['Date'], df['Price_3'])
# plt.plot(df['Date'], df['Price_4'], linestyle='--')

sns.lineplot(x=df['Date'], y=df['Price_4'])
plt.title('Sample Line Plotting')
plt.xlabel('Date')
plt.ylabel('Price($)')
plt.grid(True)
plt.legend(['2x', '1.5x', '1.75x'])
plt.show()
```

```python
x = np.linspace(0,10, 1000)
y = np.sin(x)
y1 = np.sin(2*x)
y3 = np.sin(4*x)
y4 = np.array([0.40 for x in range(0,1000)])

plt.figure(figsize=(15,4))
plt.plot(x,y)
# plt.plot(x,y1)
# plt.plot(x,y4,linestyle="--", color='r')
plt.plot(x,y3, linestyle="--", color='g')
plt.title('Sine Funtion')
plt.xlabel('Radians')
plt.ylabel('Sin(x)')
plt.grid(True)
plt.show()
```

```python
x = np.linspace(0,10, 1000)
y = np.exp(x)

plt.figure(figsize=(6,4))
plt.plot(x,y)
plt.title('Exponential Funtion')
plt.xlabel('t')
plt.ylabel('e^t')
plt.grid(True)
plt.show()
```

```python
x = np.arange(-100,100)
y = x ** 2

plt.figure(figsize=(6,4))
plt.plot(x,y)
plt.title('Power Funtion')
plt.xlabel('t')
plt.ylabel('x^2')
plt.grid(True)
plt.show()
```

```python
x = np.arange(-100,100)
y = x ** 3

plt.figure(figsize=(6,4))
plt.plot(x,y)
plt.title('Power Funtion')
plt.xlabel('t')
plt.ylabel('x^2')
plt.grid(True)
plt.show()
```

```python
# Simple Harmonic Oscillation

t = np.linspace(0, 10, 1000)
plt.figure(figsize=(14,4))
y = 1 * np.exp(-0.2 * t) * np.cos(2 * np.pi*t)
y2 = 1 * np.exp(-0.05 * t) * np.cos(2 * np.pi*t)
plt.plot(t, y, label='Damped Oscillation', color='crimson')
plt.plot(t, y2, label='Damped Oscillation', color='blue', linestyle='--')
plt.xlabel('Time (t)')
plt.ylabel('Displacement (y)')
plt.title('Damped Harmonic Motion')
plt.grid(True)
plt.legend()
plt.show()
```

## Bar Plot

```python
df_group = df.groupby('Color')['Size (cm)'].mean()
df_group = pd.DataFrame(data=df_group)
df_group.reset_index(inplace=True)
df_group
```

```python
plt.bar(x=df_group['Color'], height=df_group['Size (cm)'])
plt.show()
```

```python
df_group = df.groupby('Variety')['Size (cm)'].mean()
df_group = pd.DataFrame(data=df_group)
df_group.reset_index(inplace=True)
# df_group

plt.figure(figsize=(30,4))
sns.barplot(x=df_group['Variety'], y=df_group['Size (cm)'])
plt.show()
```

```python
plt.figure(figsize=(5,10))
sns.barplot(y=df_group['Variety'], x=df_group['Size (cm)'])
plt.title('Average Size by Variety')
plt.show()
```
